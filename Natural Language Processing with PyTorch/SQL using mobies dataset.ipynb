Import libraries

```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('movies.sqlite')

tables = pd.read_sql("""SELECT *
FROM sqlite_master
WHERE type = 'table';""", conn)

tables
```

```python
directors = pd.read_sql("""SELECT *
FROM directors;""", conn)

directors
```

```python
movies = pd.read_sql("""SELECT *
FROM movies;""", conn)
movies
```

```python
pd.read_sql("""SELECT original_title, overview, tagline
FROM movies;""", conn)
```

LIMIT

```python
pd.read_sql("""SELECT original_title, overview, tagline
FROM movies LIMIT 5;""", conn)
```

DISTINCT

```python
pd.read_sql("""SELECT DISTINCT title
FROM movies;""", conn)
```

WHERE

```python
pd.read_sql("""SELECT popularity
FROM movies WHERE popularity > 40;""", conn)
```

ORDER BY

```python
pd.read_sql("""SELECT vote_average
FROM movies ORDER BY vote_average ASC;""", conn)
```

COUNT

```python
pd.read_sql("""SELECT COUNT(id)
FROM movies;""", conn)
```

```python
pd.read_sql("""SELECT AVG(vote_average)
AS "VOTE AVERAGE"
FROM movies;""", conn)
```

```python
pd.read_sql("""SELECT gender, COUNT(*) AS[COUNT]
FROM directors
GROUP BY gender
ORDER BY 2 DESC; """, conn)
```

```python
pd.read_sql("""SELECT COUNT(*)
FROM directors
WHERE uid IS NULL;""", conn)
```
